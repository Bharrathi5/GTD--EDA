# **Project Summary -**

The Global Terrorism Database (GTD) is a comprehensive un classified database containing more than 200,000 records and details of terrorist incidents worldwide since 1970. It is maintained by the National Consortium for the Study of Terrorism and Responses to Terrorism (START) at the University of Maryland, United States. The GTD defines a terrorist attack as the threatened or actual use of illegal force and violence by a nonstate actor to attain a political, economic, religious, or social goal through fear, coercion, or
intimidation. GTD allows a researcher/government organisation/risk analyst to explore the patterns and trends behind global terrorism activities and take a data-driven approach to better understand the complex phenomenon of terrorism as it is a sensitive global concern on human lives. Using Exploratory Data Analysis (EDA), one can uncover hidden insights and reveal the key factors influencing terrorist activities around the world through visually appealing graphs and plots using various tools available on python libraries specially built for data visualization tasks.

## 3. ***Data Wrangling***

### Data Wrangling Code

**Columns filtering:** There are too many columns to account for. Hence it is necessary to filter the most relevant and important variables for analysis.

```python
# Write your code to make your dataset analysis ready.
#Filtering the necessary/selected columns for analysis
gtd = terror.filter(['iyear','imonth','iday',
  'country_txt','region_txt','provstate','city','latitude','longitude',
  'attacktype1_txt','targtype1_txt','targsubtype1_txt','multiple','success','weaptype1_txt','weapsubtype1_txt',
  'gname','suicide','individual','nperps','claimed'
  'motive','crit1','crit2','crit3','ransom','ransomamt','ransompaid','natlty1_txt',
  'nkill','nwound','property','propextent_txt','propvalue','ishostkid'
   ], axis=1)
```

```python
#updated shape of data
print("Updated shape of the data after filtering:",gtd.shape)
```

```python
#renaming complex column names for better understanding
gtd.rename(columns={'iyear':'year','imonth':'month','iday':'day',
  'country_txt':'country','region_txt':'region','provstate':'state',
  'attacktype1_txt':'attacktype','targtype1_txt':'targettype','targsubtype1_txt':'targetsubtype',
  'weaptype1_txt':'weapontype','weapsubtype1_txt':'weaponsubtype','natlty1_txt':'nationality',
  'propextent_txt':'propextent'},inplace=True)
```

#### **15 - Pair Plot**

```python
# Pair Plot visualization code
pair=gtd.filter(['year','country','success',
  'attacktype','targetype','weapontype',
  'gname','nationality','nkill','nwound',
  'ncasuality'], axis=1).sort_values(by=['ncasuality'],ascending=False)
fildf=pair.head(60).sort_values(by=['year'])
sns.pairplot(fildf,hue='success')
plt.show()
```

##### 1. Why did you pick the specific chart?

***To plot multiple pairwise bivariate distributions in a dataset, pairplot() function can be used.Each variable is plotted both in the rows and columns, showing the relationships between the variables. Here, the since the dataset is large, i have filtered the most important variables in consideration to make the pairplot a visible and interpretable one.The diagonal line shows the distribution of values in that variable using a histogram. Each other cell shows the relationship as a scatterplot between the two variables at their intersection.***
